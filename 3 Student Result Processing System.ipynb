```python
# Function to process student result
def process_result(marks):
    total = 0
    
    # Loop to calculate total
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    # Check pass or fail
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"

# Sample marks list
student_marks = [60, 45, 70, 50, 40]

# Call function
average, result = process_result(student_marks)

print("Average Marks:", average)
print("Result:", result)
```
