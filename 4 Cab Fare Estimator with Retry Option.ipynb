```python
# Function to calculate cab fare
def calculate_fare(distance, peak_hour):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    # Add 25% extra during peak hours
    if peak_hour == "yes":
        fare += fare * 0.25
    
    return fare

# Main program with retry option
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ").lower()
        
        total_fare = calculate_fare(distance, peak_hour)
        
        print(f"Total Fare: â‚¹{total_fare}")
        
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using Cab Fare Estimator ðŸš–")
            break
            
    except ValueError:
        print("Please enter a valid distance.")
```
