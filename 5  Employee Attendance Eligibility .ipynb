```python
# Function to check attendance eligibility
def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0
    
    # Loop to count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    attendance_percentage = (present_days / total_days) * 100
    
    if attendance_percentage >= 75:
        return attendance_percentage, "Eligible"
    else:
        return attendance_percentage, "Not Eligible"

# Sample attendance list
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

# Call function
percentage, result = check_attendance(attendance)

print("Attendance Percentage:", percentage, "%")
print("Status:", result)
```
